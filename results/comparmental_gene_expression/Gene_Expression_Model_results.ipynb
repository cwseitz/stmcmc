```python
file2full = np.load('gene_expression_2_mcmc_full.npz')
file3full = np.load('gene_expression_3_mcmc_full.npz')
file4full = np.load('gene_expression_4_mcmc_full.npz')

file2tuned = np.load('gene_expression_2_mcmc_tuned.npz')
file3tuned = np.load('gene_expression_3_mcmc_tuned.npz')
file4tuned = np.load('gene_expression_4_mcmc_tuned.npz')

new2 = np.load('newold2.npz')
new3 = np.load('newold3.npz')
new4 = np.load('newold4.npz')
```

```python
tevals2full = file2full['tevals']
tevals3full = file3full['tevals']
tevals4full = file4full['tevals']
tevals2tuned = file2tuned['tevals']
tevals3tuned = file3tuned['tevals']
tevals4tuned = file4tuned['tevals']

times2full = file2full['times']
times3full = file3full['times']
times4full = file4full['times']
times2tuned = file2tuned['times']
times3tuned = file3tuned['times']
times4tuned = file4tuned['times']

beta2full = file2full['beta']
beta3full = file3full['beta']
beta4full = file4full['beta']
beta2tuned = file2tuned['beta']
beta3tuned = file3tuned['beta']
beta4tuned = file4tuned['beta']

model2tuned = file2tuned['modelused']
model3tuned = file3tuned['modelused']
model4tuned = file4tuned['modelused']

model2full = max(model2tuned)*np.ones(beta2full.shape)
model3full = max(model3tuned)*np.ones(beta3full.shape)
model4full = max(model4tuned)*np.ones(beta4full.shape)
```

```python
copymaxs = np.linspace(10, 120, 6, dtype=int)
genespace = [2, 3, 4]

methods = ['full', 'tuned']
method_names = ['Full-fidelity', 'Tuned IT-Bridge']

par_names2 = ['k_{{01}}', 'k_{{10}}', 'r_1', 'k_{{trans}}', '\\gamma']
par_names3 = ['k_{{01}}', 'k_{{12}}', 'k_{{10}}', 'k_{{21}}', 'r_1', 'r_2', 'k_{{trans}}', '\\gamma']
par_names4 = ['k_{{01}}', 'k_{{12}}', 'k_{{23}}', 'k_{{10}}', 'k_{{21}}', 'k_{{32}}', 'r_1', 'r_2', 'r_3', 'k_{{trans}}', '\\gamma']

method_colors = ['Red', 'Magenta']
method_ls = ['--', 'dotted']
```

```python
lsest_u = 0;
lsest_l = np.zeros(len(beta2full)-1)
lsest_u = np.zeros(len(beta2full)-1)
for inc in range(0,(len(beta2full)-1)):
    db = beta2full[inc+1]-beta2full[inc]
    
    w_l = loglike2full[:,inc]-max(loglike2full[:,inc])
    w_u = loglike2full[:,inc+1]-max(loglike2full[:,inc+1])
    lsest_l[inc] = np.log(np.mean(np.exp(db*w_l))) + db*max(loglike2full[:,inc])
    lsest_u[inc] = np.log(np.mean(np.exp(db*w_u))) + db*max(loglike2full[:,inc+1])
    
evidence2full = [np.sum(lsest_l), 1.0/2.0*(np.sum(lsest_u)+np.sum(lsest_l)), np.sum(lsest_u)]
print(evidence2full)
print((np.sum(lsest_l)-np.sum(lsest_u))/2.0)
times2full[-1]
```
