```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filedatfull = np.load('repressilator_mcmc_full.npz')
filedatmult = np.load('repressilator_mcmc_multifi.npz')
```

```python
timesfull = filedatfull['times']
timesmult = filedatmult['times']
betafull = filedatfull['beta']
betamult = filedatmult['beta']
```

```python
plt.loglog(betafull,timesfull)
plt.loglog(betamult,timesmult)
```

```python
thetafull = filedatfull['theta']
thetamult = filedatmult['theta']
```

```python
for ipara in range(0,12):
    plt.semilogx(betafull,np.mean(thetafull[:,ipara,:],axis=0),'r')
    plt.semilogx(betafull,np.mean(thetafull[:,ipara,:],axis=0)+np.std(thetafull[:,ipara,:],axis=0),'m')
    plt.semilogx(betafull,np.mean(thetafull[:,ipara,:],axis=0)-np.std(thetafull[:,ipara,:],axis=0),'m')
    plt.semilogx(betamult,np.mean(thetamult[:,ipara,:],axis=0),'b')
    plt.semilogx(betamult,np.mean(thetamult[:,ipara,:],axis=0)+np.std(thetamult[:,ipara,:],axis=0),'c')
    plt.semilogx(betamult,np.mean(thetamult[:,ipara,:],axis=0)-np.std(thetamult[:,ipara,:],axis=0),'c')
    plt.show()
```
