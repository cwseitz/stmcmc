```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filedatfull = np.load('il1b_mcmc_full.npz')
filedatmult = np.load('il1b_mcmc_multifi.npz')
filedattuned = np.load('il1b_mcmc_tuned.npz')
filedatbridge = np.load('il1b_mcmc_bridge.npz')
```

```python
tevalsfull = filedatfull['tevals']
tevalsmult = filedatmult['tevals']
tevalstuned = filedattuned['tevals']
tevalsbridge = filedatbridge['tevals']

timesfull = filedatfull['times']
timesmult = filedatmult['times']
timestuned = filedattuned['times']
timesbridge = filedatbridge['times']

betafull = filedatfull['beta']
betamult = filedatmult['beta']
betatuned = filedattuned['beta']
betabridge = filedatbridge['beta']
```

```python
plt.loglog(betafull,timesfull)
plt.loglog(betamult,timesmult)
plt.loglog(betatuned,timestuned)
plt.loglog(betabridge,timesbridge)
plt.legend(('Full','Multifi','Multifi with Beta Tuned','ESS Bridged'))
#plt.legend(('Multifi','Multifi with Beta Tuned','ESS Bridged'))
```

```python
thetafull = filedatfull['theta']
thetamult = filedatmult['theta']
thetatuned = filedattuned['theta']
thetabridge = filedatbridge['theta']
```

```python
for ipara in range(0,11):
    #plt.semilogx(betafull,np.mean(thetafull[:,ipara,:],axis=0),'r')
    plt.semilogx(betafull,np.mean(thetafull[:,ipara,:],axis=0)+np.std(thetafull[:,ipara,:],axis=0),'b')
    plt.semilogx(betafull,np.mean(thetafull[:,ipara,:],axis=0)-np.std(thetafull[:,ipara,:],axis=0),'b')
    #plt.fill_between(betafull,np.mean(thetafull[:,ipara,:],axis=0)-np.std(thetafull[:,ipara,:],axis=0),np.mean(thetafull[:,ipara,:],axis=0)+np.std(thetafull[:,ipara,:],axis=0),alpha=0.5)
    
    #plt.semilogx(betamult,np.mean(thetamult[:,ipara,:],axis=0),'b')
    plt.semilogx(betamult,np.mean(thetamult[:,ipara,:],axis=0)+np.std(thetamult[:,ipara,:],axis=0),'orange')
    plt.semilogx(betamult,np.mean(thetamult[:,ipara,:],axis=0)-np.std(thetamult[:,ipara,:],axis=0),'orange')
    #plt.fill_between(betamult,np.mean(thetamult[:,ipara,:],axis=0)-np.std(thetamult[:,ipara,:],axis=0),np.mean(thetamult[:,ipara,:],axis=0)+np.std(thetamult[:,ipara,:],axis=0),alpha=0.5)

    #plt.semilogx(betatuned,np.mean(thetatuned[:,ipara,:],axis=0),'g')
    plt.semilogx(betatuned,np.mean(thetatuned[:,ipara,:],axis=0)+np.std(thetatuned[:,ipara,:],axis=0),'g')
    plt.semilogx(betatuned,np.mean(thetatuned[:,ipara,:],axis=0)-np.std(thetatuned[:,ipara,:],axis=0),'g')
    #plt.fill_between(betatuned,np.mean(thetatuned[:,ipara,:],axis=0)-np.std(thetatuned[:,ipara,:],axis=0),np.mean(thetatuned[:,ipara,:],axis=0)+np.std(thetatuned[:,ipara,:],axis=0),alpha=0.5)
  
    #plt.semilogx(betabridge,np.mean(thetabridge[:,ipara,:],axis=0),'k')
    plt.semilogx(betabridge,np.mean(thetabridge[:,ipara,:],axis=0)+np.std(thetabridge[:,ipara,:],axis=0),'r')
    plt.semilogx(betabridge,np.mean(thetabridge[:,ipara,:],axis=0)-np.std(thetabridge[:,ipara,:],axis=0),'r')
    #plt.fill_between(betabridge,np.mean(thetabridge[:,ipara,:],axis=0)-np.std(thetabridge[:,ipara,:],axis=0),np.mean(thetabridge[:,ipara,:],axis=0)+np.std(thetabridge[:,ipara,:],axis=0),alpha=0.5)
    plt.xscale('log')
  
    plt.legend(('Full','Multifi','Multifi with Beta Tuned','ESS Bridged'))
    #plt.legend(('Multifi','Multifi with Beta Tuned','ESS Bridged'))
    plt.show()
```
